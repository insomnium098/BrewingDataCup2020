```python
import geopandas
import matplotlib.pyplot as plt
import pandas as pd
```

# Fronteras de MÃ©xico


```python
mex=geopandas.read_file("mexican-states/mexican-states.shp")
```

```python
ruta = 'ubicaciones.csv'
ubicaciones = pd.read_csv(ruta,
                header = 0)
```

```python
gdf = geopandas.GeoDataFrame(
    ubicaciones, geometry=geopandas.points_from_xy(ubicaciones.lon, ubicaciones.lat))
```

# Mapa sin zoom

```python
fig,ax = plt.subplots(1,1,figsize=(10,8))
mex.plot(ax=ax,color = 'lightgray', edgecolor = 'black')
gdf.plot(ax = ax, marker='*', color='red', markersize=5)
plt.show()
```

# Mapa con zoom

```python
fig,ax = plt.subplots(1,1,figsize=(10,8))
mex.plot(ax=ax,color = 'lightgray', edgecolor = 'black')
gdf.plot(ax = ax, marker='*', color='red', markersize=5)
minx, miny, maxx, maxy = -98.05, 20.40, -98.35, 20.63
ax.set_xlim(minx, maxx)
ax.set_ylim(miny, maxy)
plt.show()
```

# Ya con los clusters definidos

```python
ruta = 'zona_mapa1.csv'
zona1 = pd.read_csv(ruta,
                header = 0)
ruta = 'zona_mapa2.csv'
zona2 = pd.read_csv(ruta,
                header = 0)
ruta = 'zona_mapa3.csv'
zona3 = pd.read_csv(ruta,
                header = 0)
ruta = 'zona_mapa4.csv'
zona4 = pd.read_csv(ruta,
                header = 0)
ruta = 'zona_mapa5.csv'
zona5 = pd.read_csv(ruta,
                header = 0)
ruta = 'zona_mapa6.csv'
zona6 = pd.read_csv(ruta,
                header = 0)
```

```python
gdf1 = geopandas.GeoDataFrame(
    zona1, geometry=geopandas.points_from_xy(zona1.lon, zona1.lat))
gdf2 = geopandas.GeoDataFrame(
    zona2, geometry=geopandas.points_from_xy(zona2.lon, zona2.lat))
gdf3 = geopandas.GeoDataFrame(
    zona3, geometry=geopandas.points_from_xy(zona3.lon, zona3.lat))
gdf4 = geopandas.GeoDataFrame(
    zona4, geometry=geopandas.points_from_xy(zona4.lon, zona4.lat))
gdf5 = geopandas.GeoDataFrame(
    zona5, geometry=geopandas.points_from_xy(zona5.lon, zona5.lat))
gdf6 = geopandas.GeoDataFrame(
    zona6, geometry=geopandas.points_from_xy(zona6.lon, zona6.lat))
```

# Mapa sin zoom

```python
fig,ax = plt.subplots(1,1,figsize=(10,8))
mex.plot(ax=ax,color = 'lightgray', edgecolor = 'black')
gdf1.plot(ax = ax, marker='*', color='red', markersize=5)
gdf2.plot(ax = ax, marker='*', color='green', markersize=5)
gdf3.plot(ax = ax, marker='*', color='yellow', markersize=5)
gdf4.plot(ax = ax, marker='*', color='orange', markersize=5)
gdf5.plot(ax = ax, marker='*', color='blue', markersize=5)
gdf6.plot(ax = ax, marker='*', color='purple', markersize=5)
plt.show()
```

# Mapa con zoom

```python
fig,ax = plt.subplots(1,1,figsize=(10,8))
mex.plot(ax=ax,color = 'lightgray', edgecolor = 'black')
gdf1.plot(ax = ax, marker='*', color='red', markersize=5, label = "Zona 1")
gdf2.plot(ax = ax, marker='*', color='green', markersize=5, label = "Zona 2")
gdf3.plot(ax = ax, marker='*', color='yellow', markersize=5, label = "Zona 3")
gdf4.plot(ax = ax, marker='*', color='orange', markersize=5, label = "Zona 4")
gdf5.plot(ax = ax, marker='*', color='blue', markersize=5, label = "Zona 5")
gdf6.plot(ax = ax, marker='*', color='purple', markersize=5, label = "Zona 6")

minx, miny, maxx, maxy = -98.05, 20.40, -98.35, 20.63
ax.set_xlim(minx, maxx)
ax.set_ylim(miny, maxy)

plt.legend()
plt.show()
```
